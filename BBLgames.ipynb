```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def game_plot(G):    
    T1 = np.amax(G.iloc[:,0:4], axis=1)
    T2 = np.amax(G.iloc[:,4:8], axis=1)
    stats = G.iloc[:,8:]
    over3_rolling = []
    for i in range(int(len(stats.iloc[:,1])/2)-2):
        over3_rolling.append((stats.iloc[i,1] + stats.iloc[i+1,1] + stats.iloc[i+2,1])/3)
    over3_rolling.insert(0,0)
    over3_rolling.insert(1,stats.iloc[1,1])
    over3_rolling[2] = (stats.iloc[1,1]+stats.iloc[2,1])/2
    for i in range(int(len(stats.iloc[:,1])/2)-2):
        over3_rolling.append((stats.iloc[i+20,1] + stats.iloc[i+20+1,1] + stats.iloc[i+20+2,1])/3)
    over3_rolling.insert(20,stats.iloc[20,1])
    over3_rolling.insert(21,(stats.iloc[21,1]+stats.iloc[20,1])/2)
    plt.plot(over3_rolling, '--k')
    plt.plot(stats.iloc[:,1], color='y', marker='*', markersize=10.0, markerfacecolor='k', markevery=[i for i, x in enumerate(stats.iloc[:,0] > 0) if x])
    for i in [i for i, x in enumerate(stats.iloc[:,0] == 2) if x]:
        plt.annotate(s='2', xy=(float(i)-0.4, 0.0))
    for i in [i for i, x in enumerate(stats.iloc[:,0] > 2) if x]:
        plt.annotate(s='3+', xy=(float(i)-0.4, 0.0))
    plt.ylabel("Runs")
    plt.title("Runs each over with wickets and rolling average")
    plt.grid()
    plt.show()
    plt.plot(T1)
    plt.plot(T2, color='r')
    x = np.zeros(40)+1
    plt.plot(x, color='k', linestyle='--')
    plt.ylim(0.4,1.6)
    plt.xlabel("Overs")
    plt.ylabel("Odds")
    plt.title("Both teams odds")
    plt.grid()
    plt.show()
    print(max(G.iloc[0,0:4]))
    print(max(G.iloc[0,4:8]))
```

```python
def short_game_plot(G,i1,i2):    
    T1 = np.amax(G.iloc[:,0:4], axis=1)
    T2 = np.amax(G.iloc[:,4:8], axis=1)
    stats = G.iloc[:,8:]
    over3_rolling = []
    for i in range(i1-2):
        over3_rolling.append((stats.iloc[i,1] + stats.iloc[i+1,1] + stats.iloc[i+2,1])/3)
    over3_rolling.insert(0,0)
    over3_rolling.insert(1,stats.iloc[1,1])
    over3_rolling[2] = (stats.iloc[1,1]+stats.iloc[2,1])/2
    for i in range(i2-2):
        over3_rolling.append((stats.iloc[i+i1,1] + stats.iloc[i+i1+1,1] + stats.iloc[i+i1+2,1])/3)
    over3_rolling.insert(i1,stats.iloc[i1,1])
    over3_rolling.insert(i1+1,(stats.iloc[i1+1,1]+stats.iloc[i1,1])/2)
    plt.plot(over3_rolling, '--k')
    plt.plot(stats.iloc[:,1], color='y', marker='*', markersize=10.0, markerfacecolor='k', markevery=[i for i, x in enumerate(stats.iloc[:,0] > 0) if x])
    for i in [i for i, x in enumerate(stats.iloc[:,0] == 2) if x]:
        plt.annotate(s='2', xy=(float(i)-0.4, 0.0))
    for i in [i for i, x in enumerate(stats.iloc[:,0] > 2) if x]:
        plt.annotate(s='3+', xy=(float(i)-0.4, 0.0))
    plt.ylabel("Runs")
    plt.title("Runs each over with wickets and rolling average")
    plt.grid()
    plt.show()
    plt.plot(T1)
    plt.plot(T2, color='r')
    x = np.zeros(i1+i2+1)+1
    plt.plot(x, color='k', linestyle='--')
    plt.ylim(0.4,1.6)
    plt.xlabel("Overs")
    plt.ylabel("Odds")
    plt.title("Both teams odds")
    plt.grid()
    plt.show()
    print("Innings 1:", i1, "overs")
    print("Innings 2:", i2, "overs")
    print(max(G.iloc[0,0:4]))
    print(max(G.iloc[0,4:8]))
```

```python
G1 = pd.read_csv("StvPS10.csv", header=None)
game_plot(G1)
```
